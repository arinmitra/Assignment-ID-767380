# In summary:

<b><u>Label Encoding</u></b> is better for <u><i>ordinal data</i></u> or when you want to reduce the dimensionality created by many categories. <b><u>One-Hot Encoding</u></b> is better for <u><i>nominal data</i></u> to avoid imposing an artificial order and is generally preferred for most machine learning algorithms when dealing with nominal categorical variables.
In your specific case with 'sex', 'smoker', and 'region', there is no inherent order to these categories (e.g., 'male' is not greater than 'female'). Therefore, One-Hot Encoding might be a more appropriate choice than Label Encoding for these features if you were to use algorithms sensitive to numerical relationships. However, Label Encoding is simpler to implement and can be used as a starting point or with certain algorithms that handle categorical features differently.

Therefore as suggested by Google Gemini and in keeping with the nature of the data, which seems to be of nominal nature, we will go by the preffered encoding as in method 2, i.e. <u><i><b>One-Hot Encoding</b></i></u>, and take that dataframe for any further actions as required.

````python
# Function to plot bar charts for quantiles of charges by a categorical column
def plot_categorical_charges_quantiles(df, categorical_col_encoded, title, quantiles_to_get=[0.25, 0.5, 0.75]):
    """
    Plots bar charts of specified quantiles for 'charges' grouped by a categorical column.

    Args:
        df (pd.DataFrame): The input DataFrame.
        categorical_col_encoded (str): The name of the encoded categorical column.
        title (str): The title for the plots.
        quantiles_to_get (list): A list of quantiles to plot.
    """
    quantiles = df.groupby(categorical_col_encoded)['charges'].quantile(quantiles_to_get).unstack()

    plt.figure(figsize=(10, 6))
    ax = quantiles.plot(kind='bar', color=['skyblue', 'orange'], legend=True)

    plt.title(f'Quantiles of Charges by {title}', fontsize=14)
    plt.xlabel(f'{title.split(" (")[0]} Category', fontsize=12)
    plt.ylabel('Charges', fontsize=12)
    plt.xticks(rotation=0) # Keep labels horizontal
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add values on top of the bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%.2f', label_type='edge', fontsize=8)


    plt.tight_layout()
    plt.show()

# Plot quantiles for sex_male
plot_categorical_charges_quantiles(df_encoded, 'sex_male', 'Sex (0: Female, 1: Male)')

# Plot quantiles for smoker_yes
plot_categorical_charges_quantiles(df_encoded, 'smoker_yes', 'Smoker Status (0: No, 1: Yes)')

# Plot quantiles for region_northwest
plot_categorical_charges_quantiles(df_encoded, 'region_northwest', 'Region (0: Not Northwest, 1: Northwest)')

# Plot quantiles for region_southeast
plot_categorical_charges_quantiles(df_encoded, 'region_southeast', 'Region (0: Not Southeast, 1: Southeast)')

# Plot quantiles for region_southwest
plot_categorical_charges_quantiles(df_encoded, 'region_southwest', 'Region (0: Not Southwest, 1: Southwest)')
````
